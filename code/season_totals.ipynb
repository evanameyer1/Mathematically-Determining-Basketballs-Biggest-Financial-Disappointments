```python
import requests
from bs4 import BeautifulSoup
import uuid
import difflib
import html5lib
import time
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
pd.options.mode.chained_assignment = None 
```

```python
player_index = pd.read_csv('../updated_datasets/player_index.csv')
player_index = player_index[player_index.columns[1:]]
player_index
```

```python
def season_totals(start_year, end_year):
    df_list = []
    for i in range(start_year + 1, end_year + 2):
        var_name = f'season_totals_{i - 1}'
        r = requests.get(f'https://www.basketball-reference.com/leagues/NBA_{i}_totals.html')
        if r.status_code == 200:
            soup = BeautifulSoup(r.content, 'html.parser')
            table = soup.find('table')
            test = pd.read_html(str(table))[0]
            
            #table cleaning
            test.columns = test.columns.str.lower()
            test = test[test['player'] != 'Player'].drop(columns=['rk'])
            test = test.replace('\*','',regex=True)
            test.fillna(0, inplace=True)
            globals()[var_name] = test.merge(player_index)
            df_list.append(globals()[var_name])
            print(f'{var_name} worked!')
            
        else:
            print(f'ERROR {r.status_code}')
        
        time.sleep(30)
        
    return df_list
```

```python
df_list = season_totals(1996, 2021)
```

```python
def remove_duplicates(tables):
    cleaned_df_list = []
    for i in range(len(tables)):
        years = [year for year in range(1996, 2022)]
        var_name = f'season_totals_{years[i]}'
        globals()[var_name] = tables[i][tables[i].duplicated(subset=['player'], keep='first') == False].reset_index(drop=True)
        cleaned_df_list.append(globals()[var_name])
        globals()[var_name].to_csv(f'../updated_datasets/season_totals/{var_name}.csv')
        print(f'{var_name} is done!')
    return cleaned_df_list
```

```python
def load_data():
    loaded_list = []
    for i in [year for year in range(1990, 2022)]:
        var_name = f'season_totals_{i}'
        table = pd.read_csv(f'../updated_datasets/season_totals/{var_name}.csv')
        globals()[var_name] = table[table.columns[1:]]
        globals()[var_name]['year'] = i
        globals()[var_name].to_csv(f'../updated_datasets/season_totals/{var_name}.csv')
        loaded_list.append(var_name)
    return loaded_list
```
