```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
import uuid
```

```python
nba_salaries = pd.DataFrame({'Player':[], 'Salary':[], 'Date':[]})
nba_salaries
```

```python
year = '2000-2001/'
salaries_2000 = pd.read_html(requests.get(f'https://hoopshype.com/salaries/players/{year}').text)[0].drop(
    columns = ['Unnamed: 0', f'{year[:4]}/{year[7:9]}(*)'])

salaries_2000.rename(columns={f'{year[:4]}/{year[7:9]}':'Salary'}, inplace=True)
salaries_2000[salaries_2000.columns[1:]] = salaries_2000[salaries_2000.columns[1:]].replace('[\$,]', '', regex=True).astype(float)
salaries_2000['Year'] = year[:4]

for name in salaries_2000['Player'].unique():
    salaries_2000.loc[salaries_2000['Player'] == name, 'UUID'] = uuid.uuid4()

salaries_2000.to_csv(f'../original_datasets/salaries_{year[:4]}.csv')
nba_salaries = pd.concat([nba_salaries, salaries_2000]).reset_index(drop=True)
salaries_2000
```
