```python
def make_table(num):
    table_list = []
    for i in range(num + 1):
        #define unique variables for each table
        year = f'{2000 + i}-{2001 + i}/'
        var_name = f'salaries_{2000 + i}'
        
        #grab the data
        globals()[var_name]= pd.read_html(requests.get(f'https://hoopshype.com/salaries/players/{year}').text)[0].drop(
        columns = ['Unnamed: 0', f'{year[:4]}/{year[7:9]}(*)'])
        
        #clean the dataset
        globals()[var_name].rename(columns={f'{year[:4]}/{year[7:9]}':'Salary'}, inplace=True)
        globals()[var_name][globals()[var_name].columns[1:]] = globals()[var_name][globals()[var_name].columns[1:]].replace(
            '[\$,]', '', regex=True).astype(float)
        globals()[var_name]['Year'] = year[:4]
        
        #merge and save
        globals()[var_name] = pd.merge(left=globals()[var_name], right=player_index, how='left', on='Player')
        globals()[var_name].to_csv(f'../updated_datasets/salary_data/{var_name}.csv')
        table_list.append(var_name)
    return table_list
```
