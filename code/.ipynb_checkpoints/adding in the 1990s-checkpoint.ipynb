```python
new_players = []
```

```python
player_index = pd.read_csv('../updated_datasets/player_index.csv')
player_index = player_index[player_index.columns[1:]]
player_index
```

```python
def make_table(num):
    table_list = []
    df_list = []
    for i in range(num + 1):
        #define unique variables for each table
        year = f'{1990 + i}-{1991 + i}/'
        var_name = f'salaries_{1990 + i}'
        
        #grab the data
        globals()[var_name] = pd.read_html(requests.get(f'https://hoopshype.com/salaries/players/{year}').text)[0].drop(
        columns = ['Unnamed: 0', f'{year[:4]}/{year[7:9]}(*)'])
        
        #clean the dataset
        globals()[var_name].rename(columns={f'{year[:4]}/{year[7:9]}':'Salary'}, inplace=True)
        globals()[var_name][globals()[var_name].columns[1:]] = globals()[var_name][globals()[var_name].columns[1:]].replace(
            '[\$,]', '', regex=True).astype(float)
        globals()[var_name]['Year'] = year[:4]
        
        #add the players to the player_index
        for name in globals()[var_name]['Player']:
            if name in player_index:
                continue
            else:
                new_players.append(name)
        
        #merge and save
        df_list.append(globals()[var_name])
        table_list.append(var_name)
    
    return table_list
```

```python
make_table(9)
```

```python
new_df = pd.DataFrame(new_players, columns = ['Player']).drop_duplicates().reset_index(drop=True)
new_df["UUID"] = None
new_df
```

```python
player_index = pd.concat([player_index, new_df]).reset_index(drop=True)
for name in player_index['Player'].unique():
    if player_index.loc[player_index['Player'] == name, 'UUID'].isnull().any():
        player_index.loc[player_index['Player'] == name, 'UUID'] = uuid.uuid4()
    else:
        continue
player_index.dropna().drop_duplicates()
player_index
```
