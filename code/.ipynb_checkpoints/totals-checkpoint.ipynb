```python
def load_csvs(string):
    table_list = []
    for i in range(22):
        year = 2000 + i
        var_name = f'{string}_{year}'
        globals()[var_name] = pd.read_csv(f'../updated_datasets/{string}_data/{string}_{year}.csv')
        globals()[var_name] = globals()[var_name][globals()[var_name].columns[1:]]
        table_list.append(var_name)
    return table_list
```

```python
def yearly_totals(table1, table2):
    table_list = []
    for i in range(22):
        year = 2000 + i
        var_name1 = f'{table1}_{year}'
        var_name2 = f'{table2}_{year}'
        merge_name = f'totals_{year}'
        globals()[merge_name] = pd.merge(left=globals()[var_name1],
                                         right=globals()[var_name2][['UUID', 'Salary']], how='left', on='UUID')
        globals()[merge_name].dropna(inplace=True)
        table_list.append(merge_name)
    return table_list
```
